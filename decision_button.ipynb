# 決定ボタンのノートブック

このノートブックは「決定」ボタンを作成し、ボタンを押すと「ボタンが押されました」と表示します。


```python
# ipywidgetsをインポート
import ipywidgets as widgets
from IPython.display import display

# 出力用のエリアを作成
output = widgets.Output()

# 決定ボタンを作成
decision_button = widgets.Button(
    description='決定',
    button_style='info',
    tooltip='決定ボタンをクリックしてください',
    icon='check',
    layout=widgets.Layout(width='200px', height='50px')
)

# ボタンが押されたときの処理
def on_button_clicked(b):
    with output:
        output.clear_output()
        print("ボタンが押されました")

# ボタンにイベントリスナーを設定
decision_button.on_click(on_button_clicked)

# ボタンと出力エリアを表示
display(decision_button, output)
```
